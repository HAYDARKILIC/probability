## 1. The Basic Counting Principle

**Rule:** If an experiment consists of $r$ distinct stages and the $i$-th stage has $n_i$ possible outcomes, the total number of outcomes is:

$$n_1 \times n_2 \times \cdots \times n_r = \prod_{i=1}^{r} n_i$$

### Example 1.1 — Team Formation
How many distinct teams consisting of 1 girl and 1 boy can be formed from a group of 5 girls and 2 boys?

```python
girls = ['G1', 'G2', 'G3', 'G4', 'G5']
boys  = ['B1', 'B2']

# Basic counting principle
total = len(girls) * len(boys)
print(f"Number of girls : {len(girls)}")
print(f"Number of boys  : {len(boys)}")
print(f"Total teams     : {len(girls)} × {len(boys)} = {total}")

print("\nAll possible teams:")
team_no = 1
teams = []
for g in girls:
    for b in boys:
        print(f"  {team_no:2d}. ({g}, {b})")
        teams.append((g, b))
        team_no += 1
```

```python
# Visualization: Cartesian product table
fig, ax = plt.subplots(figsize=(9, 4))
ax.set_xlim(0, len(boys) + 0.5)
ax.set_ylim(0, len(girls) + 0.5)
ax.set_xticks(range(1, len(boys) + 1))
ax.set_yticks(range(1, len(girls) + 1))
ax.set_xticklabels(boys,  fontsize=12)
ax.set_yticklabels(girls, fontsize=12)
ax.set_xlabel("Boys",  fontsize=12)
ax.set_ylabel("Girls", fontsize=12)
ax.set_title(f"Basic Counting Principle: {len(girls)} × {len(boys)} = {total} distinct teams",
             fontsize=13, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.4)

colors = plt.cm.Set2(np.linspace(0, 1, len(teams)))
for idx, (g, b) in enumerate(teams):
    xi = boys.index(b)  + 1
    yi = girls.index(g) + 1
    ax.scatter(xi, yi, s=400, color=colors[idx], zorder=3, edgecolors='black', linewidths=0.8)
    ax.text(xi, yi, f"{g}\n{b}", ha='center', va='center', fontsize=7, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Example 1.2 — License Plate Count (Generalized Rule)
How many distinct license plates of the format **LLL-DDD** (3 letters + 3 digits) can be produced?

```python
n_letters = 26   # A-Z
n_digits  = 10   # 0-9

# Generalized product rule: 26^3 × 10^3
stages = [n_letters, n_letters, n_letters,
          n_digits,  n_digits,  n_digits]

total_plates = math.prod(stages)

print("Stage      | Number of Choices")
print("-" * 30)
names = ["1st Letter", "2nd Letter", "3rd Letter", "1st Digit", "2nd Digit", "3rd Digit"]
for name, n in zip(names, stages):
    print(f"{name:<12} | {n}")
print("-" * 30)
print(f"\nTotal = 26 × 26 × 26 × 10 × 10 × 10")
print(f"      = 26³ × 10³")
print(f"      = {26**3:,} × {10**3:,}")
print(f"      = {total_plates:,} distinct plates")
```

## 2. Permutations

### 2.1 — Basic Permutation
The number of **all** orderings of $n$ distinct objects:
$$P(n) = n!$$

### 2.2 — Permutation with Repeated Objects
If among $n$ objects there are $n_1$ alike, $n_2$ alike, …, $n_r$ alike:
$$P = \frac{n!}{n_1!\, n_2!\, \cdots\, n_r!}$$

### 2.3 — Permutation of $r$ objects chosen from $n$
$$P(n, r) = \frac{n!}{(n-r)!}$$

```python
# Helper functions
def factorial(n):
    """Computes n!"""
    return math.factorial(n)

def perm(n, r=None):
    """Permutation of r objects from n: P(n,r) = n! / (n-r)!"""
    if r is None:
        r = n
    if r > n:
        return 0
    return factorial(n) // factorial(n - r)

def perm_with_repeats(object_list):
    """Permutation with repeated objects: n! / (n1! * n2! * ...)"""
    n = len(object_list)
    counter = Counter(object_list)
    denominator = math.prod(factorial(v) for v in counter.values())
    return factorial(n) // denominator

print("Permutation functions defined ✓")
```

### Example 2.1 — Permutations of the Letters a, b, c

```python
letters = ['a', 'b', 'c']
n = len(letters)

# List all permutations
all_perms = list(itertools.permutations(letters))

print(f"n = {n} distinct letters → {n}! = {factorial(n)} permutations")
print("\nAll permutations:")
for i, p in enumerate(all_perms, 1):
    print(f"  {i}. {''.join(p)}")

print(f"\nVerification: number of permutations found via itertools = {len(all_perms)}")
if len(all_perms) == factorial(3):
    print(f"Formula check: 3! = {factorial(3)}  ✓")
else:
    print(f"Formula check: expected={factorial(3)}, found={len(all_perms)}  ✗")
```

### Example 2.2 — Permutations of the Word "FACEBOOK" (with Repetition)

```python
word = list("FACEBOOK")
n = len(word)

print(f"Word: {''.join(word)}")
print(f"Number of letters (n): {n}")

# Letter frequencies
freq = Counter(word)
print("\nLetter frequencies:")
for letter, count in sorted(freq.items()):
    print(f"  '{letter}': {count}")

# Calculations
all_distinct = factorial(n)
with_repeats  = perm_with_repeats(word)

print(f"\n(a) As if all distinct: 8! = {all_distinct:,}")
print(f"(b) Correcting for repeated 'o': 8! / 2! = {all_distinct:,} / {factorial(2)} = {with_repeats:,}")

# Verification via itertools
actual = len(set(itertools.permutations(word)))
print(f"\nVerification (unique permutations via itertools): {actual:,}")
print("Formula correct ✓" if with_repeats == actual else "Formula incorrect ✗")
```

### Example 2.3 — Choosing 2 Numbers from {1,2,3,4,5,6} — P(6,2)

```python
S = [1, 2, 3, 4, 5, 6]
n, r = 6, 2

selections = list(itertools.permutations(S, r))

print(f"Set: {S}")
print(f"Number of elements selected: r = {r}")
print(f"\nFormula: P({n},{r}) = {n}! / ({n}-{r})! = {factorial(n)} / {factorial(n-r)} = {perm(n,r)}")
print(f"\nAll ordered selections ({len(selections)} total):")

# Grid display
row = []
for i, s in enumerate(selections):
    row.append(str(s))
    if len(row) == 6:
        print("  " + "   ".join(row))
        row = []
if row:
    print("  " + "   ".join(row))

print(f"\nNote: (1,2) and (2,1) → DIFFERENT because order matters!")
```

## 3. Combinations

**Order does not matter!** The number of ways to choose $r$ objects from $n$:

$$\binom{n}{r} = C(n,r) = \frac{n!}{(n-r)!\, r!}$$

**Permutations vs Combinations:**
$$\binom{n}{r} = \frac{P(n,r)}{r!}$$

Because the $r!$ different orderings of the selected $r$ objects are all counted as identical.

```python
def combination(n, r):
    """C(n,r) = n! / ( (n-r)! * r! )"""
    if r < 0 or r > n:
        return 0
    return math.comb(n, r)   # Python 3.8+ built-in

print("Combination examples:")
for (n, r) in [(4,2), (6,2), (52,5), (10,3)]:
    c = combination(n, r)
    p = perm(n, r)
    print(f"  C({n},{r}) = {n}!/({r}!·{n-r}!) = {c:>10,}   [P({n},{r})={p:,} / {r}!={factorial(r)}]")
```

### Example 3.1 — Handshakes: Choosing 2 from {1,2,3,4,5,6}

```python
people = [1, 2, 3, 4, 5, 6]
n, r = 6, 2

combinations_list = list(itertools.combinations(people, r))

print(f"Number of handshakes in a group of 6 = C(6,2) = {combination(n,r)}")
print(f"Formula: 6! / (4! · 2!) = {factorial(6)} / ({factorial(4)}·{factorial(2)}) = {combination(n,r)}")
print(f"\nAll handshakes:")
for i, (a, b) in enumerate(combinations_list, 1):
    print(f"  {i:2d}. Person {a} ↔ Person {b}")

print(f"\nNote: In permutations there were P(6,2)={perm(6,2)} selections.")
print(f"Since (1,2) = (2,1) for handshakes: {perm(6,2)} / 2! = {combination(n,r)}")
```

```python
# Handshake network graph
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: handshake network
ax = axes[0]
n_people = 6
angles = np.linspace(0, 2 * np.pi, n_people, endpoint=False)
x_pos = np.cos(angles)
y_pos = np.sin(angles)

for (i, j) in itertools.combinations(range(n_people), 2):
    ax.plot([x_pos[i], x_pos[j]], [y_pos[i], y_pos[j]],
            'b-', alpha=0.3, linewidth=1.2)

node_colors = plt.cm.Set1(np.linspace(0, 1, n_people))
for i in range(n_people):
    ax.scatter(x_pos[i], y_pos[i], s=500, color=node_colors[i],
               zorder=5, edgecolors='black', linewidths=1.5)
    ax.text(x_pos[i] * 1.18, y_pos[i] * 1.18, f"Person {i+1}",
            ha='center', va='center', fontsize=10, fontweight='bold')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title(f"C(6,2) = {combination(6,2)} Handshakes",
             fontsize=12, fontweight='bold')

# Right: Permutation vs Combination comparison
ax2 = axes[1]
ns_range = range(2, 11)
r_fixed = 2
perm_vals = [perm(n, r_fixed) for n in ns_range]
comb_vals = [combination(n, r_fixed) for n in ns_range]

x = np.array(list(ns_range))
width = 0.35
ax2.bar(x - width/2, perm_vals, width, label=f'P(n,{r_fixed}) — Order Matters',
        color='#E07B54', alpha=0.85, edgecolor='black', linewidth=0.5)
ax2.bar(x + width/2, comb_vals, width, label=f'C(n,{r_fixed}) — Order Irrelevant',
        color='#5B8DB8', alpha=0.85, edgecolor='black', linewidth=0.5)

ax2.set_xlabel('n (total number of elements)', fontsize=11)
ax2.set_ylabel('Count', fontsize=11)
ax2.set_title(f'P(n,{r_fixed}) vs C(n,{r_fixed}) — for r={r_fixed}',
              fontsize=12, fontweight='bold')
ax2.legend(fontsize=9)
ax2.set_xticks(x)
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

### Example 3.2 — Poker Hand: Choosing 5 Cards from 52

```python
n_cards   = 52
hand_size = 5

total_hands = combination(n_cards, hand_size)
print(f"Total number of poker hands: C(52,5) = {total_hands:,}")
print(f"Formula: 52! / (47! · 5!) = (52×51×50×49×48) / (5×4×3×2×1)")
print(f"       = {52*51*50*49*48:,} / {5*4*3*2*1} = {total_hands:,}")

# Selected special hand types
print("\n--- Special Hands ---")

# Flush: 5 cards of the same suit (4 suits × C(13,5))
flush = 4 * combination(13, 5)
print(f"Flush (5 cards same suit)            : {flush:>8,}  → probability: {flush/total_hands:.5f}")

# Full House: 3 of a kind + 2 of a kind → C(13,1)×C(4,3) × C(12,1)×C(4,2)
full_house = combination(13,1)*combination(4,3) * combination(12,1)*combination(4,2)
print(f"Full House (3+2 of same rank)        : {full_house:>8,}  → probability: {full_house/total_hands:.5f}")

# Four of a Kind: C(13,1)×C(4,4) × C(48,1)
four_of_a_kind = combination(13,1) * combination(4,4) * 48
print(f"Four of a Kind (4 same rank)         : {four_of_a_kind:>8,}  → probability: {four_of_a_kind/total_hands:.5f}")
```

### Properties of Combinations and the Binomial Theorem

```python
print("=" * 50)
print("Properties of Combinations")
print("=" * 50)

# Property 1
for n in [5, 8, 10]:
    assert combination(n, 1) == n
    assert combination(n, n) == 1
print(f"✓ Property 1: C(n,1)=n, C(n,n)=1  →  Verified")

# Property 2: C(n,r) = C(n, n-r)
for n in [6, 9]:
    for r in range(n+1):
        assert combination(n, r) == combination(n, n-r)
print(f"✓ Property 2: C(n,r) = C(n,n-r)  →  Verified")
print(f"   Example: C(9,3) = {combination(9,3)} = C(9,6) = {combination(9,6)}")

# Property 3: Pascal's Rule
for n in range(2, 10):
    for r in range(1, n+1):
        assert combination(n,r) == combination(n-1,r) + combination(n-1,r-1)
print(f"✓ Property 3 (Pascal's Rule): C(n,r)=C(n-1,r)+C(n-1,r-1)  →  Verified")
print(f"   Example: C(6,3) = C(5,3)+C(5,2) = {combination(5,3)}+{combination(5,2)} = {combination(6,3)}")
```

```python
# Pascal's Triangle Visualization
rows = 9
fig, ax = plt.subplots(figsize=(11, 6))
ax.set_xlim(-0.5, rows - 0.5)
ax.set_ylim(-0.5, rows - 0.5)
ax.axis('off')
ax.set_title("Pascal's Triangle — C(n,r) values", fontsize=13, fontweight='bold')

max_val = combination(rows - 1, (rows - 1) // 2)
cmap = plt.cm.Blues

for n in range(rows):
    for r in range(n + 1):
        val = combination(n, r)
        x = r - n / 2 + (rows - 1) / 2
        y = rows - 1 - n
        intensity = 0.15 + 0.75 * (val / max_val) ** 0.4
        circle = plt.Circle((x, y), 0.42, color=cmap(intensity), zorder=2)
        ax.add_patch(circle)
        fontsize = 10 if val < 1000 else 7
        ax.text(x, y, str(val), ha='center', va='center',
                fontsize=fontsize, fontweight='bold',
                color='white' if intensity > 0.5 else 'black', zorder=3)
        ax.text(x - 0.5, y + 0.28, f"C({n},{r})",
                ha='center', va='center', fontsize=5, color='gray', zorder=3)

plt.tight_layout()
plt.show()
print("The sum of any two adjacent values in a row gives the value directly below (Pascal's Rule).")
```

### Binomial Theorem
$$
(a+b)^n = \sum_{k=0}^{n} \binom{n}{k} a^k b^{n-k}
$$

```python
a, b = sp.symbols('a b')

print("Binomial Theorem — Expansions:")
print("-" * 45)
for n in range(1, 7):
    expansion = sp.expand((a + b)**n)
    expansion_str = str(expansion).replace('**', '^').replace('*', '·')
    print(f"(a+b)^{n} = {expansion_str}")

# Numerical verification
print("\nNumerical Verification:")
for n in [3, 5, 8]:
    # (a+b)^n = sum C(n,k) a^k b^(n-k)
    a_val, b_val = 2, 3
    direct = (a_val + b_val) ** n
    total  = sum(combination(n, k) * a_val**k * b_val**(n-k) for k in range(n+1))
    print(f"  ({a_val}+{b_val})^{n} = {direct} = {total}  {'✓' if direct==total else '✗'}")
```

## 4. Multinomial Coefficients

Partitioning $n$ objects into $r$ groups (group sizes: $n_1, n_2, \ldots, n_r$, summing to $n$):

$$\binom{n}{n_1, n_2, \ldots, n_r} = \frac{n!}{n_1!\, n_2!\, \cdots\, n_r!}$$

Multinomial Theorem:
$$(a_1 + a_2 + \cdots + a_r)^n = \sum_{n_1+\cdots+n_r=n} \binom{n}{n_1, n_2, \ldots, n_r} a_1^{n_1} a_2^{n_2} \cdots a_r^{n_r}$$

```python
def multinomial(n, groups):
    """Multinomial coefficient: n! / (n1! * n2! * ... * nr!)"""
    assert sum(groups) == n, "Group sizes must sum to n!"
    denominator = math.prod(factorial(g) for g in groups)
    return factorial(n) // denominator

print("Multinomial Coefficient Examples:")
print("-" * 55)

# Police department example
n = 10
groups = [5, 2, 3]
result = multinomial(n, groups)
print(f"Example: 10 officers → [patrol:{groups[0]}, paperwork:{groups[1]}, donut shop:{groups[2]}]")
print(f"  C(10; 5,2,3) = 10! / (5!·2!·3!) = {factorial(10)} / ({factorial(5)}·{factorial(2)}·{factorial(3)})")
print(f"             = {result:,}")

print()
# Further examples
examples = [
    (6,  [2, 2, 2], "6 players into 3 equal teams"),
    (12, [4, 4, 4], "12 students into 3 equal groups"),
    (9,  [3, 3, 3], "9 projects into 3 groups"),
]
for n, grp, description in examples:
    print(f"{description}: C({n}; {grp}) = {multinomial(n, grp):,}")

# Special case r=2: binomial coefficient
print("\nSpecial case r=2 → Binomial coefficient:")
for n in [5, 8, 10]:
    r1 = n // 2
    r2 = n - r1
    ct   = multinomial(n, [r1, r2])
    binom = combination(n, r1)
    print(f"  C({n}; {r1},{r2}) = {ct} = C({n},{r1}) = {binom}  {'✓' if ct==binom else '✗'}")
```

```python
# Visualization: Police department example
n_officers = 10
duty_names  = ['Patrol (5)', 'Paperwork (2)', 'Donut Shop (3)']
duty_counts = [5, 2, 3]
duty_colors = ['#4C72B0', '#DD8452', '#55A868']

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: Pie chart
axes[0].pie(duty_counts, labels=duty_names, colors=duty_colors,
            autopct=lambda p: f'{p:.0f}%\n({round(p*n_officers/100)} officers)',
            startangle=90, textprops={'fontsize': 11},
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[0].set_title(f'10 Officers — Duty Assignment\nC(10;5,2,3) = {multinomial(10,[5,2,3]):,} ways',
                  fontsize=12, fontweight='bold')

# Right: Multinomial vs binomial comparison
ax2 = axes[1]
ns2 = range(3, 13)
# C(n, n//3, n//3, n - 2*(n//3)) approximately equal 3 groups
vals_3 = []
vals_2 = []
for n2 in ns2:
    g1 = n2 // 3
    g2 = n2 // 3
    g3 = n2 - g1 - g2
    if g1 > 0 and g2 > 0 and g3 > 0:
        vals_3.append(multinomial(n2, [g1, g2, g3]))
    else:
        vals_3.append(0)
    vals_2.append(combination(n2, n2 // 2))

x2 = np.array(list(ns2))
ax2.semilogy(x2, vals_2, 'o-', label='C(n, n/2) — 2 groups', color='#4C72B0', linewidth=2)
x2_filt = x2[[i for i, v in enumerate(vals_3) if v > 0]]
ax2.semilogy(x2_filt, [v for v in vals_3 if v > 0],
             's--', label='C(n; n/3,n/3,n/3) — 3 groups',
             color='#DD8452', linewidth=2,
             markersize=7)
ax2.set_xlabel('n (total objects)', fontsize=11)
ax2.set_ylabel('Number of Partitions (log scale)', fontsize=11)
ax2.set_title('Binomial vs Multinomial Growth', fontsize=12, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
